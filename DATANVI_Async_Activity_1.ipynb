```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 1, 13, 7, 9]
plt.plot(x, y)

plt.title('Moving Lines!')
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.show()
```

```python
x1 = [1, 2, 3, 4, 5]
y1 = [10, 1, 13, 7, 9]

x2 = [6, 1, 2, 4, 8]
y2 = [13, 8, 14, 5, 10]

plt.plot(x1, y1, label = "line 1")
plt.plot(x2, y2, label = "line 2")

plt.title('Two Moving Lines!')
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.show()
```

```python
x = [6,5,4,3,2,1]
y = [7,8,10,9,6,2]

plt.plot(x, y, color='green', linestyle='dashed', linewidth = 7,
         marker='o', markerfacecolor='purple', markersize=15)

plt.ylim(1,12)
plt.xlim(1,7)

plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('Customized Graph!')

plt.show()
```

```python
left = [6, 7, 8, 9, 10]
height = [50, 25, 70, 10, 35]

tick_label = ['six', 'seven', 'eight', 'nine', 'ten']

plt.bar(left, height, tick_label = tick_label,
        width = 1, color = ['red', 'green', 'orange', 'blue', 'violet'])

plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('A Bar Chart!')

plt.show()
```

```python
viewership = [5, 12, 55, 33, 45, 38, 60, 47, 52, 58, 49,
              20, 10, 15, 35, 29, 70, 65, 25, 40, 43]

viewership_range = (0, 80)
bins = 8

plt.hist(viewership, bins, viewership_range, color='skyblue',
         histtype='bar', rwidth=0.8)

plt.xlabel('Viewership in Millions')
plt.ylabel('Number of Shows/Movies')
plt.title('Distribution of Viewership Across Shows/Movies')

plt.show()
```

```python
x = [1,5,7,4,3,9]
y = [8,3,2,1,4,6]

plt.scatter(x, y, label= "stars", color= "yellow",
            marker= "*", s=50)

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('The Stars')
plt.legend()

plt.show()
```

```python
sports = ['baseball', 'basketball', 'soccer', 'swimming']

slices = [3, 7, 8, 6]
colors = ['r', 'y', 'g', 'b']

plt.pie(slices, labels = sports, colors=colors,
        startangle=90, shadow = True, explode = (0, 0, 0.1, 0),
        radius = 1.2, autopct = '%1.1f%%')

plt.legend()
plt.show()
```

```python
x = np.arange(0,5*(np.pi), 0.2)
y = np.cos(x)

plt.plot(x, y)
plt.show()
```
